```python
from autoscraper import AutoScraper
```

```python
amazon_url='https://www.babiesrus.co.za/shop-by-department/nappies-changing'
print(amazon_url)
```

```python
wanted_list=['Complete Clean Baby Wipes 4 x 64 Wipes','R124.90',' Babyjoy Pants','https://www.babiesrus.co.za/media/catalog/product/cache/f1f6de05b7f74ffe86c1d44f4118e994/8/8/8851111419319_imageoutofpack.png']
```

```python
scraper=AutoScraper()
result=scraper.build(amazon_url,wanted_list)
for data in range(0, 3):
    data = scraper.get_result_similar(amazon_url,grouped=True)
    print(data)
```

```python
keys = list(data.keys())
print(keys)
```

```python
scraper.set_rule_aliases({'rule_sotw':'Title', 'rule_w221':'Price', 'rule_6yaw':'Categories', 'rule_4fnk':'Images'})
```

```python
scraper.keep_rules({'rule_sotw':'Title', 'rule_w221':'Price', 'rule_6yaw':'Categories', 'rule_4fnk':'Images'})
```

```python
scraper.save('amazon-search')
```

```python
scraper.load("amazon-search")
```

```python
result = scraper.get_result_similar("https://www.babiesrus.co.za/shop-by-department/health-medical", group_by_alias=True)
```

```python
import pandas as pd
df = pd.DataFrame(dict([ (k,pd.Series(v)) for k,v in result.items() ]))
# Converting our DataFrame to CSV file
df.to_csv('baby4Us.csv')
```
