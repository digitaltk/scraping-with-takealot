```python
from autoscraper import AutoScraper
```

```python
amazon_url='https://www.hificorp.co.za/products/appliances'
print(amazon_url)
```

```python
wanted_list=['Samsung 689lt Frost Free French Door Fridge RF28K9360SG ','R66,999','https://www.hificorp.co.za/media/catalog/product/cache/74d213abfbc037213e1c69aa712be9b2/1/0/10112733_LG_601LT_SBS_FRIDGE_SILVER_GC_L247SLUV_31c5.jpg']
```

```python
scraper=AutoScraper()
result=scraper.build(amazon_url,wanted_list)
for data in range(0, 4):
    data = scraper.get_result_similar(amazon_url,grouped=True)
    print(data)
```

```python
keys = list(data.keys())
print(keys)
```

```python
scraper.set_rule_aliases({'rule_kpkf':'Title', 'rule_dlkh':'Price', 'rule_pwgb':'Image'})
```

```python
scraper.keep_rules({'rule_kpkf':'Title', 'rule_dlkh':'Price', 'rule_pwgb':'Image'})
```

```python
scraper.save('amazon-search')
```

```python
scraper.load("amazon-search")
```

```python
result = scraper.get_result_similar("https://www.hificorp.co.za/products/health-beauty", group_by_alias=True)
```

```python
import pandas as pd
df = pd.DataFrame(dict([ (k,pd.Series(v)) for k,v in result.items() ]))
# Converting our DataFrame to CSV file
df.to_csv('Hificorp.csv')
```
