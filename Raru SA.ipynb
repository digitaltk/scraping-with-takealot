```python
from autoscraper import AutoScraper
import pandas as pd
```

```python
amazon_url='https://www.bidorbuy.co.za/dailydeals'
print(amazon_url)
```

```python
wanted_list=[' Couch Marbella 2 Seater SPECIAL RUNNING','R4,999','OttomanHomeDecor','https://img.bidorbuy.co.za/image/upload/dpr_1.0,f_auto,t_btbnx/user_images/455/2488455/170823110118_Marbella%202%20Seater.jpg']
```

```python
scraper=AutoScraper()
result=scraper.build(amazon_url,wanted_list)
data = scraper.get_result_similar(amazon_url,grouped=True)
print(data)
```

```python
keys = list(data.keys())
print(keys)
```

```python
scraper.set_rule_aliases({'rule_a399':'Name', 'rule_o3sf':'Price', 'rule_au0d':'Categories','rule_t8hr':'Images'})
```

```python
scraper.keep_rules({'rule_a399':'Name', 'rule_o3sf':'Price', 'rule_au0d':'Categories','rule_t8hr':'Images'})
```

```python
scraper.save('amazon-search')
```

```python
scraper.load("amazon-search")
```

```python
result = scraper.get_result_similar("https://www.bidorbuy.co.za/static/PremiumListings.html", group_by_alias=True)
```

```python
import pandas as pd
df = pd.DataFrame(dict([ (k,pd.Series(v)) for k,v in result.items() ]))
# Converting our DataFrame to CSV file
df.to_csv('BidorBuy.csv')
```
