```python
from bs4 import BeautifulSoup
import requests
from csv import writer

url = 'https://www.incredible.co.za/specials'
page = requests.get(url)
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
lists = soup.find_all('li',class_='product-item')
```

```python
with open('incredible.csv','w', encoding='utf8',newline='') as f:
    thewriter = writer(f)
    header = ['Title', 'Price', 'Review']
    thewriter.writerow(header)
    for list in lists:
        title = list.find('a', class_='product-item-link')
        price = list.find('span', class_='price')
        review = list.find('span', class_='count')
        info = [title, price, review]
        print(info.decompose())
        thewriter.writerow(info)
```
