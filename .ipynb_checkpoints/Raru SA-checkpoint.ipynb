```python
from autoscraper import AutoScraper
```

```python
amazon_url='https://www.mrp.com/mens/shop-by-category/chino-pants'
print(amazon_url)
```

```python
wanted_list=['Skinny Fit Chino Pants','R199.99']
```

```python
scraper=AutoScraper()
result=scraper.build(amazon_url,wanted_list)
data = scraper.get_result_similar(amazon_url,grouped=True)
print(data)
```

```python
keys = list(data.keys())
print(keys)
```

```python
scraper.set_rule_aliases({'rule_sotw':'Name', 'rule_w221':'Price', 'rule_6yaw':'Images'})
```

```python
scraper.keep_rules({'rule_sotw':'Title', 'rule_w221':'Price', 'rule_6yaw':'Categories', 'rule_4fnk':'Images'})
```

```python
scraper.save('amazon-search')
```

```python
scraper.load("amazon-search")
```

```python
result = scraper.get_result_similar("https://www.mrp.com/mens/shop-by-category/pullovers", group_by_alias=False)
```

```python
import pandas as pd
df = pd.DataFrame(dict([ (k,pd.Series(v)) for k,v in result.items() ]))
# Converting our DataFrame to CSV file
df.to_csv('baby4Us.csv')
```
