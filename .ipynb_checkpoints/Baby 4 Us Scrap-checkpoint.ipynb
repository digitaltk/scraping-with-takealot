```python
from autoscraper import AutoScraper
```

```python
amazon_url='https://www.babiesrus.co.za/shop-by-department/infant-food-feeding'
print(amazon_url)
```

```python
wanted_list=['Cerelac Regular Stage 1 - 250g ','R36.90',' Nestle','https://www.babiesrus.co.za/media/catalog/product/cache/f1f6de05b7f74ffe86c1d44f4118e994/1/0/1079648_imageinpack.jpg']
```

```python
scraper=AutoScraper()
result=scraper.build(amazon_url,wanted_list)
data = scraper.get_result_similar(amazon_url,grouped=True)
print(data)
```

```python
keys = list(data.keys())
print(keys)
```

```python
scraper.set_rule_aliases({'rule_nwv5':'Title', 'rule_3s2i':'Price', 'rule_j43v':'Categories', 'rule_qev2':'Image'})
```

```python
scraper.keep_rules({'rule_nwv5':'Title', 'rule_3s2i':'Price', 'rule_j43v':'Categories', 'rule_qev2':'Image'})
```

```python
scraper.save('amazon-search')
```

```python
scraper.load("amazon-search")
```

```python
result = scraper.get_result_similar("https://www.babiesrus.co.za/shop-by-department/maternity-pregnancy", group_by_alias=True)
```

```python
import pandas as pd
df = pd.DataFrame(dict([ (k,pd.Series(v)) for k,v in result.items() ]))
# Converting our DataFrame to CSV file
df.to_csv('baby4Us2.csv')
```
