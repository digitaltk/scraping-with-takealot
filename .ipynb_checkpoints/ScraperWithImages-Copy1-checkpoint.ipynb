```python
from autoscraper import AutoScraper
```

```python
amazon_url='https://www.babyboom.co.za/product-category/baby-linen-and-clothing/'
print(amazon_url)
```

```python
wanted_list=["Baby Campcot Comforter Set Tiny Tatty Teddy","Products, Baby Bedding, Baby Linen & Clothing, Blankets, Duvet Sets, Gifting Ideas, New Products, Sleep Time",'299.99','https://www.babyboom.co.za/wp-content/uploads/2021/05/6007856110215-1-1-300x300.jpg']
```

```python
scraper=AutoScraper()
result=scraper.build(amazon_url,wanted_list)
```

```python
data = scraper.get_result_similar(amazon_url,grouped=True)
print(data)
```

```python
keys = list(data.keys())
print(keys)
```

```python
scraper.set_rule_aliases({'rule_ivo8':'Title', 'rule_4zoz':'Categories', 'rule_83bz':'Price', 'rule_36il':'Images'})
```

```python
scraper.keep_rules({'rule_ivo8':'Title', 'rule_4zoz':'Categories', 'rule_83bz':'Price', 'rule_36il':'Images'})
```

```python
scraper.save('amazon-search')
```

```python
scraper.load("amazon-search")
```

```python
result = scraper.get_result_similar("https://www.babyboom.co.za/product-catalogue/", group_by_alias=True)
```

```python
import pandas as pd
df = pd.DataFrame(dict([ (k,pd.Series(v)) for k,v in result.items() ]))
# Converting our DataFrame to CSV file
df.to_csv('Baby_Boom.csv')
```
