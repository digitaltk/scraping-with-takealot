```python
from autoscraper import AutoScraper
```

```python
amazon_url='https://www.babiesrus.co.za/shop-by-department/nappies-changing'
print(amazon_url)
```

```python
wanted_list=['Complete Clean Baby Wipes 4 x 64 Wipes','R124.90','https://www.babiesrus.co.za/media/catalog/product/cache/f1f6de05b7f74ffe86c1d44f4118e994/6/0/6009545433810_imageoutofpackimageview1front.jpg']
```

```python
scraper=AutoScraper()
result=scraper.build(amazon_url,wanted_list)
```

```python
data = scraper.get_result_similar(amazon_url,grouped=True)
print(data)
```

```python
keys = list(data.keys())
print(keys)
```

```python
scraper.set_rule_aliases({'rule_ga1t':'Title', 'rule_83bu':'Price', 'rule_et0h':'Images'})
```

```python
scraper.keep_rules({'rule_ga1t':'Title', 'rule_83bu':'Price', 'rule_et0h':'Images'})
```

```python
scraper.save('amazon-search')
```

```python
scraper.load("amazon-search")
```

```python
result = scraper.get_result_similar("https://www.babiesrus.co.za/shop-by-department", group_by_alias=True)
```

```python
import pandas as pd
df = pd.DataFrame(dict([ (k,pd.Series(v)) for k,v in result.items() ]))
# Converting our DataFrame to CSV file
df.to_csv('BabiesRUs.csv')
```
