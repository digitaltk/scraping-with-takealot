```python
from autoscraper import AutoScraper
```

```python
search = "iphone+11+silver"
amazon_url="https://www.amazon.in/s?k=iphone+11+silver&s=price-desc-rank"
print(amazon_url)
```

```python
wanted_list=["New Apple iPhone 12 Pro Max (128GB) - Silver","501"]
```

```python
scraper = AutoScraper()
result = scraper.build(amazon_url,wanted_list)
```

```python
data = scraper.get_result_similar(amazon_url,grouped=True)
print(data)
```

```python
key = list(data.keys())
print(key)
```

```python
scraper.set_rule_aliases({'rule_r0nb':'Title', 'rule_wsac':'Rating'})
```

```python
scraper.keep_rules({'rule_r0nb':'Title', 'rule_wsac':'Rating'})
```

```python
scraper.save('amazon-search')
```

```python
scraper.load("amazon-search")
```

```python
result = scraper.get_result_similar("https://www.amazon.in/s?k=playstation", group_by_alias=True)
```

```python
import pandas as pd
df = pd.DataFrame(dict([ (k,pd.Series(v)) for k,v in result.items() ]))
# Converting our DataFrame to CSV file
df.to_csv('PlayStation_Data.csv')
```
